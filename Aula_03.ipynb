```python
#Exiba a quantidade de aprovados e de reprovados!
```

```python
notas = [8, 9, 7, 6, 5, 4, 3, 2, 2, 1] 
aprovados = 0
reprovados = 0
for nota in notas:
  if(nota >= 7):
    aprovados += 1
  else:
    reprovados += 1

print('Aprovados: ', aprovados)
print('Reprovados: ', reprovados)
```

```python
#Criar um programa que lista os número de 1 a 10.
#Para isso, precisaremos de um contador:
cont = 1
numeros = []

while (cont <= 10):
  numeros.append(cont)
  cont += 1

numeros
```

```python
#Inserir nomes de ciades
#O laço acaba quando digitar 0

cidades = []
city = input('Insira o nome da cidade: ')

while(city != '0'):
  cidades.append(city)
  city = input('Insira o nome da cidade: ')

cidades
```
